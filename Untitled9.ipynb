```python
import pandas as pd
import pyreadstat

df, meta = pyreadstat.read_sav("sample_2 - Copie.sav")  # Remplace par le vrai nom si différent
df.head()  # pour voir les premières lignes
df.info()          # types de données
df.isnull().sum()  # valeurs manquantes par colonne
df.describe()      # résumé statistique (numérique)
```

```python
# Imports essentiels
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Afficher les premières lignes pour voir la structure
print("Aperçu des données :")
display(df.head())

# 2. Informations générales sur le DataFrame
print("\nInformations générales sur les colonnes :")
df.info()

# 3. Statistiques descriptives pour variables numériques
print("\nStatistiques descriptives :")
display(df.describe())

# 4. Distribution des valeurs manquantes
print("\nValeurs manquantes par colonne :")
print(df.isnull().sum())

# Visualisation des valeurs manquantes (heatmap)
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title("Carte des valeurs manquantes")
plt.show()

# 5. Distribution des variables catégorielles principales (exemple avec 'sexe' ou autre variable)
for col in df.select_dtypes(include='object').columns:
    print(f"\nDistribution de la variable catégorielle : {col}")
    display(df[col].value_counts())
    plt.figure(figsize=(6,3))
    sns.countplot(data=df, x=col)
    plt.title(f"Répartition de {col}")
    plt.show()



# 7. Matrice de corrélation pour variables numériques
plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Matrice de corrélation")
plt.show()
```
